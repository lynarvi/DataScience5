```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [120,121,122,123,124,125,126,127,128,129,130,131,132,133,134]

# fill in missing indexes in y
y_mean = np.mean(y)

y = [120,121,122,123,124,125,126,127,128,129,130,131,132,133,134, y_mean, y_mean, y_mean]

model = np.poly1d(np.polyfit(x, y, 3))
print(r2_score(y, model(x)))
line = np.linspace(1, 22, 134)

plt.scatter(x,y)
plt.plot(line, model(line))
plt.show()

# Since the value of r squared is 0.91, x and y have a very good relationship, and we can use polynomial regression.

```
